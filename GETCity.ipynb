```
import json
import requests
import numpy
import pandas as pd
```

```
import csv
results = []
rows, cols = (266, 6) 
rcount=0;
arrdata = [[0 for i in range(cols)] for j in range(rows)] 
with open("geolocation.csv") as csvfile:
    reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC) # change contents to floats
    for row in reader: # each row is a list
        results.append(row)
        
```

```
for rs in results:
  url=("https://api.bigdatacloud.net/data/reverse-geocode-client?latitude=",str(rs[1]),"&longitude=",str(rs[0]),"&localityLanguage=en")
  urlstring="".join(url)
  print(urlstring)
  response = requests.get(urlstring)
  todos = json.loads(response.text)
  arrdata[rcount][0]=rs[1]
  arrdata[rcount][1]=rs[0]
  arrdata[rcount][2]=todos["countryName"]
  arrdata[rcount][3]=todos["principalSubdivision"]
  arrdata[rcount][4]=todos["city"]
  arrdata[rcount][5]=todos["locality"]
  rcount=rcount+1

```

```

a = numpy.array(arrdata)
numpy.savetxt("Location.csv", a, delimiter=',', header="Lat,Long,country,state,city,locality", comments="")
```
